```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df_people = pd.read_csv('./Data/People.csv')
df_people = df_people.set_index('FIPS')
df_jobs = pd.read_csv('./Data/Jobs.csv')
df_jobs = df_jobs.set_index('FIPS')
df_income = pd.read_csv('./Data/Income.csv')
df_income = df_income.set_index('FIPS')

```

```python
df = pd.concat([df_people, df_jobs,df_income], axis=1)

```

```python
plt.scatter(df.Deep_Pov_All, df.MedHHInc, color='blue')
plt.xlabel('Deep poverty, 2013-17')
plt.ylabel('Median household income, 2017 (In 2017 dollars)')
plt.show()
```
